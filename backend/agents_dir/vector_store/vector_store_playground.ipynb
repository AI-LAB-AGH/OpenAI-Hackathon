```python
import os

def get_filenames(directory):
    """Get list of filenames from the specified directory"""
    filenames = []
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            filenames.append(filename)
    return filenames

files = get_filenames("../../../example-notes/")

```

```python
for filename in files:
    path = os.path.join("../../../example-notes/", filename)
    file = open(path, "rb")
    file_obj = client.files.create(
        file=file,
        purpose="assistants"
    )
    client.vector_stores.files.create(
        vector_store_id=vector_store_id,
        file_id=file_obj.id
    )
    
```

```python
vector_store = client.vector_stores.files.list(
  vector_store_id=vector_store_id
)
print_model(vector_store)
```
